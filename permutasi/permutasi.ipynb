Permutasi dan Kombinasi di Jupyter Notebook

```python
#Import Module
import itertools
```

```python
#Permutasi
#n! = n Faktorial
data = ['A','B','C','D','E']
permutasi = itertools.permutations(data)

for perm in permutasi:
    print(perm)
```

```python
data = ['A','B','C','D','E']
permutasi = itertools.permutations(data,5)
for kombinasi in permutasi:
    print(kombinasi)
```

```python
permutasi = itertools.permutations(range(30),10)
len(tuple(permutasi))
```

```python
#kombinasi sederhana
data = ['A','B','C','D','E','F','G']
```

```python
kombinasi = itertools.combinations(data,2)

for combine in kombinasi:
    print(combine)
```

```python
kombinasi = itertools.combinations(range(10),4)
len(tuple(kombinasi))
```

```python
kombinasi = itertools.combinations(range(10),4)
for comb in kombinasi:
    print(comb)
```
